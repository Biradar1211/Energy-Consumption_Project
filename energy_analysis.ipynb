```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#load dataset
df = pd.read_csv("household_energy.csv",parse_dates=["timestamp"])
df
```

```python
#Extract the time feature
df['hour'] = df['timestamp'].dt.hour
df['day'] = df['timestamp'].dt.day
df['weekday'] = df['timestamp'].dt.dayofweek
df['date'] = df['timestamp'].dt.date
df
```

```python
#basic summary
print('== Basic analytics')
print(df.info())
```

```python
print("\n=== Summary ===")
print(df.describe())
```

```python
# Grouping the DataFrame by the 'date' column (already extracted from 'timestamp')
# Then summing the 'energy_consumption' for each date to get total daily consumption
daily_consumption = df.groupby('date')['energy_consumption'].sum()

# Grouping the DataFrame by ISO calendar week number (from the 'timestamp' column)
# .dt.isocalendar().week extracts the ISO week (1 to 52/53)
# Then summing the 'energy_consumption' for each week to get total weekly consumption
weekly_consumption = df.groupby(df['timestamp'].dt.isocalendar().week)['energy_consumption'].sum()
```

```python
#visualize daily usage
plt.figure(figsize=(12,4))
plt.plot(daily_consumption.index,daily_consumption.values,marker='o')
plt.title("Daily household Energy Consumption")
plt.xlabel("Data")
plt.ylabel("Energy (khw)")
plt.grid(True)
plt.tight_layout()
plt.savefig("daily_energy_plot.png")
```

```python


# Creating a new figure for the weekly energy consumption plot with a custom size
plt.figure(figsize=(12, 4))

# Plotting the weekly energy consumption as a bar chart
plt.bar(weekly_consumption.index, weekly_consumption.values)

# Setting the title of the plot
plt.title("Weekly Household Energy Consumption")

# Labeling the X-axis as 'Date' (or 'Week Number' if that makes more sense in your data)
plt.xlabel("Date")

# Labeling the Y-axis as 'Energy Consumption (kWh)'
plt.ylabel("Energy Consumption (kWh)")

# Enabling gridlines for easier reading of values
plt.grid(True)

# Automatically adjusting subplot parameters for a neat layout
plt.tight_layout()

# Saving the plot as an image file named 'weekly_energy_plt.png'
plt.savefig("weekly_energy_plt.png")

# Optionally show the plot
plt.show()
```

```python
#forecasting with Linear regression
features = ['temperature','outside_temperature','device_usage','hour','weekday']
x=df[features]
y=df['energy_consumption']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=42,shuffle=True)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
joblib.dump(model,"forest_model.pkl")
```

```python
#root mean square error
y_pred = model.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print(f"\n Forecasting RMSE:{rmse:.2f}")
```

```python
#add forecast to full dataset
df['predicted_consumption']=model.predict(x)
df
```

```python
# Moving average over the past 6 readings (hours)
df['moving_avg'] = df['energy_consumption'].rolling(window=6, min_periods=1).mean()

# Flag as inefficient if consumption exceeds 29% above moving average
df['inefficient'] = df['energy_consumption'] > 1.29 * df['moving_avg']
```

```python
print("=== Inefficient Usage Entries ===")
ineff_df = df[df['inefficient']]
print(ineff_df[['timestamp', 'energy_consumption','moving_avg']].head())
```

```python
recommendations = []  # global list (not reset in later cells)

for _, row in ineff_df.iterrows():
    rec = []

    if row['device_usage'] > 0 and row['energy_consumption'] > 0:
        rec.append(" Consider turning off the device when not in use to save energy.")

    if row['device_usage'] > 30 and row['energy_consumption'] > 3:
        rec.append(" High usage with high outside temperature: Adjust thermostat settings.")

    if row['hour'] in [0, 1, 2, 3, 4] and row['energy_consumption'] > 2:
        rec.append(" High energy use detected at night. Investigate unnecessary consumption.")

    recommendations.extend(rec)  #  important: extend, not reset!
```

```python
# Remove duplicates
recommendations = list(set(recommendations))

# Print if any recommendations found
print("\n=== Recommendations ===")
if recommendations:
    for rec in recommendations:
        print(f"  - {rec}")
else:
    print(" No specific recommendations. Usage looks optimal!")
```
